```python
import matplotlib.pyplot as plt
import numpy as np


x1 = [0,1/6,1/6,1/6,1/6,1/6,1/6]
x = [x1]

    

def dices(N):
    
    if len(x)<N:
        y = []
        
        for w in range(len(x)+1):
            y.append(0)
        
        for a in range(len(x)+1,(len(x)+1)*6+1):
            p=0
            
            for c in range(1,7):
                if 6*len(x)>=a-c>=0:
                    p = p + 1/6 * x[len(x)-1][a-c]
            y.append(p)
            
        x.append(y)
        dices(N)

def gaus(x,N):
    m =3.5*N
    s =np.sqrt(35*N/12)
    return (1/(s*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-m)/s)**2)
         
    
def Anzahl(N):
    dices(N)
    z1 =[]
    z2 =[]
    z3 =np.linspace(len(x),6*len(x),100*len(x)**2)
    
    for i in range(len(x),len(x)*6+1):
        z1.append(i)
        
    for j in range(len(x),len(x)*6+1):
        z2.append(x[len(x)-1][j])
        
    plt.title(N)
    plt.plot(z1, z2, 'bo')
    plt.plot(z3, gaus(z3,len(x)), 'r')
    plt.grid(True)
    plt.show()
    
    
             

for i in range(1,11):
    Anzahl(i)
Anzahl(20)
Anzahl(200)
```
