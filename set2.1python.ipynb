```python
import matplotlib.pyplot as plt
import numpy as np


def factorial(n):
    i = 1
    j = 1
    while i <= n:
        j = j * i
        i = i + 1
    return j

def binomial(n,k):
    b = factorial(n)/(factorial(k)*factorial(n-k))
    return b
    
def pi(i,N):
    if (N+i)%2:
        return 0
    return binomial(N,(N+i)/2)*0.5**N


def p(i,N):
    s =np.sqrt(N/4)
    return (1/(s*np.sqrt(2*np.pi)))*np.exp(-0.5*(i/s)**2)



z1 = []
z2 = []



def show(N):
    
    for i in range(-N,N+1):
        z1.append(i)

    for j in range(0,2*N+1):
        z2.append(pi(z1[j],N))
    
    z3 = np.linspace(-N, N, 100*N)
    
    plt.title(N)
    plt.plot(z1, z2, 'bo')
    plt.plot(z3, p(z3,N), 'r')
    plt.grid(True)
    plt.show()

    z1.clear()
    z2.clear()
   
```

```python
for i in range(1,11):
    show(i)
    
show(100)
show(500)
show(1000)
```
